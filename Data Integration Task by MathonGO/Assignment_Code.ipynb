```python
# import necessary libraries
import re
import json
```

```python
# load text file
with open("Task.txt", "r") as file:
    text = file.read()
    
print(text)
```

```python
# defined function to extract each questions,options and solutions
def extract_text(text):
    questions = []
    current_question = {}
    option_number = 1
    solution_text = ""
    capturing_question_text = False  # Flag to indicate whether we are capturing question text
    capturing_solution_text = False  # Flag to indicate whether we are capturing solution text

    for line in text.split("\n"):
        #print("Processing line:", line) 
        if re.match(r'^\s*Question ID:|\\section\*{Question ID:', line):
            #print("Found question ID line",line)
            if current_question:
                if "questionId" in current_question: # Check if questionId is present
                    current_question["solutionText"] = solution_text.strip().replace("\\\\", "\\")    # Convert '\\\\' to '\\' in solutionText
                    current_question["questionText"] = current_question["questionText"].rstrip('\n')  # Remove the last '\n' from questionText
                    #print("questionText:", current_question.get("questionText"))  
                    #print("solutionText:", current_question.get("solutionText")) 
                    questions.append(current_question)
                current_question = {}
                solution_text = ""
                option_number = 1     # Reset option number for each new question
            current_question["questionNumber"] = len(questions) + 1
            current_question["questionId"] = int(re.search(r"\d+", line).group())
            capturing_question_text = True    # Start capturing question text
            capturing_solution_text = False
        elif re.match(r'^\s*\(A\)|\(B\)|\(C\)|\(D\)', line):
            #print("Found option line:", line.strip())
            capturing_question_text = False    # Stop capturing question text once options start
            current_question.setdefault("options", []).append({"optionNumber": option_number, "optionText": line[4:].strip(), "isCorrect": False})
            option_number += 1
        elif re.match(r'Answer \(([A-D])\)|\\section\*{Answer \(([A-D])\)}', line):
            #print("Found answer line:", line.strip()) 
            match = re.search(r'Answer \(([A-D])\)|\\section\*{Answer \(([A-D])\)}', line)
            if match:
                correct_option = match.group(1) or match.group(2)
                if current_question:
                    for option in current_question["options"]:
                        option["isCorrect"] = option["optionNumber"] == ord(correct_option) - ord('A') + 1  #Converts the correct option letter (e.g., 'A', 'B', 'C', 'D') into its Unicode code point integer value.
        elif line.startswith("Sol."):
            #print("Found solution line:", line.strip()) 
            capturing_solution_text = True   # Start capturing solution text
            solution_text += line[len("Sol."):].strip()  # Capture text immediately after 'Sol.'
        elif capturing_solution_text:
            if line.strip() == "":
                if solution_text.strip() != "":  # Add '\n' only if there is content in solution_text
                    solution_text += "\n"
            else:
                solution_text += line + " "
        elif capturing_question_text:
            if current_question.get("questionText") is None:
                current_question["questionText"] = ""
            if line.strip() != "":
                current_question["questionText"] += line.replace("\\", "\\\\").replace("\\\\", "\\").replace(" \\\\ ", " \\ ").strip() + "\n"

    if current_question and "questionId" in current_question:
        current_question["solutionText"] = solution_text.strip().replace("\\\\", "\\")
        current_question["questionText"] = current_question["questionText"].rstrip('\n')  # Remove the last '\n' from questionText
        questions.append(current_question)

    return questions
```

```python
# Dumping the final text into "Json" file

#def convert_to_json(questions):
#    return json.dumps(questions, indent=2)

questions = extract_text(text)
json_data = json.dumps(questions, indent=2)  #convert_to_json(questions)
with open("Final_output.json", "w") as file:
    file.write(json_data)
```
